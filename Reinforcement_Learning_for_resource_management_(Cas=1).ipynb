```python
gamma=0.9
def value_iteration(S,A,P,R):
    V={s:0 for s in S}
    optimal_policy={s:0 for s in S}
    while True :
        oldV = V.copy()
        for s in S:
            Q={}
            for a in A[s]:
                Q[a]=R(s,a)+gamma*sum(P(s_next,s,a)*oldV[s_next] for s_next in S)
            V[s]=max(Q.values())
            if V[s] >= oldV[s]:
              oldV[s] = V[s]
            optimal_policy[s]=max(Q,key=Q.get)
        if all(V[s]==oldV[s] for s in S):
           break
    return V,optimal_policy           
print(value_iteration(Etats, Actions,Prob_Transition,Reward))
```
